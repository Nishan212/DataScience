```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
testdf = pd.read_csv('X_Test.csv')
traindf = pd.read_csv('X_Train.csv')
payers = pd.read_csv('payers.csv')
nonPayers = pd.read_csv('nonpayers.csv')

dfs = [testdf, traindf]
for df in dfs:
    df.set_index(['Loan_ID'], inplace = True)
```
