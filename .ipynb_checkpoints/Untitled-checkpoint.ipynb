```python
import pandas as pd
```

```python
traindf = pd.read_csv('train.csv')
traindf.set_index('Loan_ID', inplace = True)
```

```python
traindf.replace({'Gender' : 'Male', 'Married' : 'Yes', 'Education' : 'Graduate'}, 1, inplace = True) 
traindf.replace({'Gender' : 'Female', 'Married' : 'No', 'Education' : 'Not Graduate'}, 0, inplace = True) 
```

```python
traindf.drop(['CoapplicantIncome', 'Dependents'], axis = 1, inplace = True)
```

```python
loanPayer = traindf[traindf['Loan_Status'] == 'Y']
nonLoanPayer = traindf[traindf['Loan_Status'] == 'N']
```

```python
loanPayer.drop(['Loan_Status'], axis = 1, inplace = True)
nonLoanPayer.drop(['Loan_Status'], axis = 1, inplace = True)
```

```python
loanPayer.to_csv('loanPayer.csv')
```

```python
nonLoanPayer.to_csv('nonLoanPayer.csv')
```
